```python
from pyspark.sql import SparkSession

# Initialize Spark session with S3 configuration
spark = SparkSession.builder \
    .appName("UploadToS3") \
    .master("local[*]") \
    .config("spark.jars.packages", "org.apache.hadoop:hadoop-aws:3.3.4,com.amazonaws:aws-java-sdk-bundle:1.12.426") \
    .config("spark.hadoop.fs.s3a.impl", "org.apache.hadoop.fs.s3a.S3AFileSystem") \
    .config("spark.hadoop.fs.s3a.aws.credentials.provider", "com.amazonaws.auth.DefaultAWSCredentialsProviderChain") \
    .config("spark.driver.extraJavaOptions", "-Djava.security.manager=allow") \
    .config("spark.executor.extraJavaOptions", "-Djava.security.manager=allow") \
    .getOrCreate()

# Load local CSV file
df = spark.read.option("header", "true").csv("../data/bangkok_traffy.csv")

# Define S3 destination path
s3_path = "s3a://traffy-troffi/traffy/fondue/bangkok_traffy.parquet"

# Write DataFrame to S3 in Parquet format
df.write.mode("overwrite").parquet(s3_path)

print("Upload complete.")
```
